```python
# Importando o pacote Pandas como 'pd'
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
# Importando o dataset completo de 'nps' (notebook2)
nps = pd.read_parquet('../data/stage/3m/customers_nps_churn_palive.parquet')
```

```python
df_loyalty = pd.read_parquet('../data/raw/3m/acelera_prd_loyaltytransactions.parquet')
```

```python
df_loyalty.loc[:,'consumer_id'] = pd.to_numeric(df_loyalty['consumer_id'], errors='coerce', downcast='integer')
```

```python
df_loyalty.head()
```

```python
df_nps_loyalty = pd.merge(nps, df_loyalty, on='consumer_id', how='left', suffixes=('_nps', '_loyalty'))
```

```python
df_nps_loyalty.to_parquet('../data/stage/3m/customers_loyalty.parquet')
```
