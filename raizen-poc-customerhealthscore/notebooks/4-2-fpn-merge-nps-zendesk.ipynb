### Converting Dtypes to merge DataFrames

```python
df_nps['cpf'] = df_nps['cpf'].astype('int').astype('str')
```

```python
df_nps['cpf'] = df_nps['cpf'].apply(lambda x: x.zfill(11)).astype('object')
```

### Merging DataFrames and checking results

```python
df_customers_zendesk = pd.merge(df_nps, df_zendesk, on='cpf', how='left')
```
