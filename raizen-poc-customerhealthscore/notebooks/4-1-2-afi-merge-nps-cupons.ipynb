# Merge ('nps' + 'Vouchers')
Para essa mescla será necessário 'transaction_id' e 'ref_date' como primary keys: transaction_id, reward_id, store_id

```python
# Como essas são as chaves primarias, precisamos nos certificar de que estão no mesmo dtype
cupons['transaction_id'] = pd.to_numeric(cupons['transaction_id'], errors = 'coerce')
cupons['reward_id'] = pd.to_numeric(cupons['reward_id'], errors = 'coerce')
cupons['store_id'] = pd.to_numeric(cupons['store_id'], errors = 'coerce')
```

```python
columns = [
    'voucher_id', 'origin_consumer_id', 'target_consumer_id', 'store_id',
    'reward_id', 'discount_amount', 'expires_at', 'transaction_id',
    'redeemed_at', 'redeem_status', 'created_at', 'updated_at',
    'count_voucher', 'partner_voucher_id', 'order_use', 'voucher_data'
]
cupons = cupons[columns]
```

```python
cupons = cupons[cupons['transaction_id'].notna()]
```

```python
cupons.loc[:,'transaction_id'] = cupons['transaction_id'].astype('int')
cupons.loc[:,'reward_id'] = cupons['reward_id'].astype('int')
```

```python
nps.shape
```

```python
# Merge tabelas nps + cupons
merge_table = pd.merge(nps, cupons, on=['transaction_id'], how='left', suffixes=('_transactions', '_vouchers'))
```

```python
merge_table.shape
```

# Salvando em parquet

```python
#Salvando em Parquet
merge_table.to_parquet('../data/stage/3m/customers_transactions_vouchers.parquet')
```
